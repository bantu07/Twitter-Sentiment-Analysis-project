```python
tw = []
def func1(user_name,date_id):
    auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
    auth.set_access_token(access_key,access_secret)
    api = tweepy.API(auth,wait_on_rate_limit=True)
    num = 300
    tweets1 = tweepy.Cursor(api.search,q=user_name,lang='en',since=date_id).items(15000)
    all_tweets = [tweet.text for tweet in tweets1]
    for i in all_tweets:
        tw.append(i)
    print(tw)
```
